```julia
using DataStructures
using Graphs

function all_simple_paths(
    g::AbstractGraph{T}, u::T, vs; cutoff::T=nv(g) - one(T)
) where {T<:Integer}
    vs = vs isa Set{T} ? vs : Set{T}(vs)
    return SimplePathIterator(g, u, vs, cutoff)
end

# iterator over all simple paths from `u` to `vs` in `g` of length less than `cutoff`
struct SimplePathIterator{T<:Integer,G<:AbstractGraph{T}}
    g::G
    u::T       # start vertex
    vs::Set{T} # target vertices
    cutoff::T  # max length of resulting paths
end

function Base.show(io::IO, spi::SimplePathIterator)
    print(io, "SimplePathIterator{", typeof(spi.g), "}(", spi.u, " → ")
    if length(spi.vs) == 1
        print(io, only(spi.vs))
    else
        print(io, '[')
        join(io, spi.vs, ", ")
        print(io, ']')
    end
    print(io, ')')
    return nothing
end
Base.IteratorSize(::Type{<:SimplePathIterator}) = Base.SizeUnknown()
Base.eltype(::SimplePathIterator{T}) where {T} = Vector{T}

mutable struct SimplePathIteratorState{T<:Integer}
    stack::Stack{Tuple{T,T}} # used to restore iteration of child vertices: elements are ↩
    # (parent vertex, index of children)
    visited::Stack{T}         # current path candidate
    queued::Vector{T}         # remaining targets if path length reached cutoff
    self_visited::Bool        # in case `u ∈ vs`, we want to return a `[u]` path once only
end
function SimplePathIteratorState(spi::SimplePathIterator{T}) where {T<:Integer}
    stack = Stack{Tuple{T,T}}()
    visited = Stack{T}()
    queued = Vector{T}()
    push!(visited, spi.u)    # add a starting vertex to the path candidate
    push!(stack, (spi.u, one(T))) # add a child node with index 1
    return SimplePathIteratorState{T}(stack, visited, queued, false)
end

function _stepback!(state::SimplePathIteratorState) # updates iterator state.
    pop!(state.stack)
    pop!(state.visited)
    return nothing
end

# iterates to the next simple path in `spi`, according to a depth-first search
function Base.iterate(
    spi::SimplePathIterator{T}, state::SimplePathIteratorState=SimplePathIteratorState(spi)
) where {T<:Integer}
    while !isempty(state.stack)
        if !isempty(state.queued) # consume queued targets
            target = pop!(state.queued)
            result = vcat(reverse(collect(state.visited)), target)
            if isempty(state.queued)
                _stepback!(state)
            end
            return result, state
        end

        parent_node, next_child_index = first(state.stack)
        children = outneighbors(spi.g, parent_node)
        if length(children) < next_child_index
            _stepback!(state) # all children have been checked, step back
            continue
        end

        child = children[next_child_index]
        next_child_index_tmp = pop!(state.stack)[2]                      # move child ↩ 
        push!(state.stack, (parent_node, next_child_index_tmp + one(T))) # index forward
        child in state.visited && continue

        if length(state.visited) == spi.cutoff
            # collect adjacent targets if more exist and add them to queue
            rest_children = Set(children[next_child_index:end])
            state.queued = collect(
                setdiff(intersect(spi.vs, rest_children), Set(state.visited))
            )

            if isempty(state.queued)
                _stepback!(state)
            end
        else
            result = if child in spi.vs
                vcat(reverse(collect(state.visited)), child)
            else
                nothing
            end

            # update state variables
            push!(state.visited, child) # move to child vertex
            if !isempty(setdiff(spi.vs, state.visited)) # expand stack until all targets are found
                push!(state.stack, (child, one(T))) # add the child node as a parent for next iteration
            else
                pop!(state.visited) # step back and explore the remaining child nodes
            end

            if !isnothing(result) # found a new path, return it
                return result, state
            end
        end
    end

    # special-case: when `vs` includes `u`, return also a 1-vertex, 0-length path `[u]`
    if spi.u in spi.vs && !state.self_visited
        state.self_visited = true
        return [spi.u], state
    end
end
```

```julia
using Random
using StatsBase  # Needed for sample()

function new_initial_solution(model)
    nNodes = model["nNodes"]
    uav = model["uav"]
    tru = model["tru"]
    L = model["L"]
    Du = model["Du"]
    
    # Generating the TruckRoute
    TruckRoute = zeros(Int, tru, nNodes+1)
    TruckRoute[:, 1] .= 1  # initial depot
    nodes = collect(2:nNodes)

    while !isempty(nodes)
        for i in 1:tru
            if !isempty(nodes)
                no = rand(1:length(nodes))  # randomly select a node
                sot = findfirst(TruckRoute[i, :] .== 0)  # find the first zero in the row
                if sot !== nothing
                    TruckRoute[i, sot] = nodes[no]
                    deleteat!(nodes, no)  # update nodes
                end
            end
        end
    end

    for i in 1:tru
        last_nonzero = findlast(!iszero, TruckRoute[i, :])
        if last_nonzero !== nothing && last_nonzero < size(TruckRoute, 2)
            TruckRoute[i, last_nonzero + 1] = 1  # final depot
        end
    end

    element = findall(TruckRoute .> 1)
    
    fu = sample(1:length(element), min(uav, length(element)), replace=false)

    n = Int[]
    for i in fu
        push!(n, TruckRoute[element[i][1], element[i][2]])  # Fix indexing for CartesianIndex
    end

    DroneRoute = zeros(Int, nNodes + 2, nNodes + 2, uav)
    u = randperm(uav)

    inserted = Int[]
    launchland = Int[]

    cc = 0
    for ii in 2:nNodes
        yes = 0
        if cc == uav
            break
        end

        if ii in launchland
            continue
        end

        row, col = Tuple(findfirst(==(ii), TruckRoute))

        push!(inserted, ii)
        Tr = copy(TruckRoute)
        Tr[row, col] = 0
        launch = filter(x -> x != 1, Tr[Tr .!= 0])

        for in_node in inserted
            filter!(x -> x != in_node, launch)
        end

        for i in launch
            if yes == 1
                break
            end

            land = vcat(TruckRoute[row, col], TruckRoute[row, col+1:findlast(!iszero, TruckRoute[row, :])])
            filter!(x -> x != ii, land)

            for j in land
                if Du[i, ii] + Du[ii, j] < L && i != j
                    uu = sample(1:length(u), 1, replace=false)[1]
                    DroneRoute[1, 1, u[uu]] = i
                    DroneRoute[1, 2, u[uu]] = ii
                    DroneRoute[1, 3, u[uu]] = j
                    append!(launchland, [i, j])
                    u = filter(x -> x != u[uu], u)
                    yes += 1
                    if yes == 1
                        if col < size(TruckRoute, 2)
                            TruckRoute[row, col:end-1] = TruckRoute[row, col+1:end]
                            TruckRoute[row, end] = 0  # Clear last column to prevent duplicates
                        end
                        cc += 1
                    end
                    
                    break
                end
            end
        end
    end

    Sol = (TruckRoute=TruckRoute, DroneRoute=DroneRoute)
    return Sol
end
```

```julia
function cost_function(model, solution, method)
 

    TruckRoute = solution["TruckRoute"]
    DroneRoute = solution["DroneRoute"]

    nNodes = model["nNodes"]
    L = model["L"]
    tru = model["tru"]
    uav = model["uav"]
    Dt = model["Dt"]
    Du = model["Du"]
    
    longestpath = 0
    penalty8 = 0
    penalty6 = 0
    penalty77 = 0
    
    #added:
    for u in 1:uav
        sa=findall(!iszero, DroneRoute[:, 1, u])
        if !isempty(sa)            
            for j in sa                
                if DroneRoute[j, 1, u] == DroneRoute[j, count(x -> !iszero(x), DroneRoute[j, :, u]), u]
                    longestpath += 100000000
                    return longestpath
                end
            end
        end
    end

    #added:
    adj = zeros(Int, nNodes + 1, nNodes + 1)
    for i in 1:nNodes        
        yechizi=findall(x -> x == i, TruckRoute)
        if !isempty(yechizi)
            for idx in yechizi
                (a, b) = Tuple(idx)  # Convert CartesianIndex to (row, col)
                if TruckRoute[a, b] == 1 && b==count(!iszero,TruckRoute[a,:])
                    #(TruckRoute[a, b + 1] == 0) && b==nNodes+1
                    continue
                elseif TruckRoute[a, b + 1] == 1
                    adj[i, nNodes + 1] = 1
                else                    
                    adj[i, TruckRoute[a, b + 1]] = 1
                end
            end
        end
        
        indices = findall(x -> x == i, DroneRoute)
        for idx in indices
            (row, col, depth) = Tuple(idx)  # Convert CartesianIndex to (row, col, depth)
        
            if col + 1 <= size(DroneRoute, 2) && DroneRoute[row, col + 1, depth] != 0
                if DroneRoute[row, col + 1, depth] == 1
                    adj[i, nNodes + 1] = 2
                else
                    adj[i, DroneRoute[row, col + 1, depth]] = 2
                end
            end
        end
    end
    #
    indegree = [count(!iszero, adj[:, i]) for i in 1:(nNodes + 1)]
    Topo = zeros(Int, nNodes + 1)
    visitnodes = 0
    savej = Int[]

    while any(x -> x >= 0, indegree)
        j = findall(x -> x == 0, indegree)
        if length(j) > 1
            for hh in j
                if !(hh in savej)
                    push!(savej, hh)
                    Topo[findfirst(==(0), Topo)] = hh
                    indegree[hh] = -1
                    visitnodes += 1
                    for k in findall(!iszero, adj[hh, :])
                        indegree[k] -= 1
                    end
                end
            end
        elseif length(j) == 1
            Topo[findfirst(==(0), Topo)] = j[1]
            push!(savej, j[1])
            indegree[j[1]] = -1
            visitnodes += 1
            for k in findall(!iszero, adj[j[1], :])
                indegree[k] -= 1
            end
        end
    end

    lable = fill(-Inf, nNodes + 1)
    lable[first(findall(!iszero, indegree))] = 0

    for ii in Topo
        for j in findall(!iszero, adj[ii, :])
            if adj[ii, j] == 1
                lable[j] = max(lable[j], lable[ii] + Dt[ii,j])
            elseif adj[ii, j] == 2
                lable[j] = max(lable[j], lable[ii] + Du[ii,j])
            end
        end
    end

    Penalty = 0
    longestpath += maximum(lable) + Penalty + penalty6 + penalty8 + penalty77
    return longestpath
end
```

```julia
function create_neighbor(model, Sol, Pr, pres, prev)
    flag = 0
    
    # Roulette wheel selection to choose a method
    r = rand()  # Generate a random number between 0 and 1
    c = cumsum(Pr)  # Cumulative sum of probabilities
    METHOD = findfirst(r .<= c)  # Find the first index where r <= c

    # Update previous and current method
    prev = pres
    pres = METHOD
    
    # Perform the selected method
    if METHOD == 1  # InsertionTd
        newSol, flag = Insertionewemtehanii(model, Sol, flag)
    elseif METHOD == 2  # InsertionDt
        newSol, flag = InsertionDt1(model, Sol, flag)
    elseif METHOD == 3  # swaptrone
        newSol, flag = swaptrone1(model, Sol, flag)
    elseif METHOD == 4  # swaptruck
        newSol, flag = swaptruck1(model, Sol, flag)
    elseif METHOD == 5  # swaptr1
        newSol, flag = swaptr1(model, Sol, flag)
    elseif METHOD == 6  # landchange2
        newSol, flag = landchange4(model, Sol, flag)
    elseif METHOD == 7  # launchange
        newSol, flag = launchange1(model, Sol, flag)
    elseif METHOD == 8  # opt2drone
        newSol, flag = opt2drone1(model, Sol, flag)
    elseif METHOD == 9  # 2opt
        newSol, flag = opt2new(model, Sol, flag)
    elseif METHOD == 10  # multivisit
        newSol, flag = multivist(model, Sol, flag)
    elseif METHOD == 11  # InsertionTT2
        newSol, flag = InsertionTT2(model, Sol, flag)
    else
        error("Invalid METHOD selected")
    end

    return newSol, flag, METHOD, pres, prev
end
```

```julia
function opt2drone1(model, Sol, flag)
    
    
    TruckRoute = Sol["TruckRoute"]
    DroneRoute = Sol["DroneRoute"]
    
    nNodes = model["nNodes"]
    ua = model["uav"]
    tru = model["tru"]
    L = model["L"]
    Du = model["Du"]

    flag = 0

    uav = []    
    for u in 1:ua
        if !isempty(findall(!iszero, DroneRoute[:, :, u]))
            satr = findall(!iszero, DroneRoute[:, 1, u])
            for i in 1:length(satr)
                if count(!iszero, DroneRoute[satr[i], :, u]) > 3
                    push!(uav, u)
                end
            end
        end
    end    
    
    if isempty(uav)
        flag = 1
        return Sol, flag
    end

    cc = 0
    aa = rand(1:length(uav))
    u = uav[aa]
    dr = findall(!iszero, DroneRoute[:, 1, u])

    for i in 1:length(dr)
        if cc == 0
            ff = DroneRoute[dr[i], 1:count(!iszero, DroneRoute[dr[i], :, u]), u]
            cost = 0
            for j in 1:length(ff)-1
                cost += Du[ff[j], ff[j+1]]
            end
            dd = ff
            dd = dd[2:end-1]  # Remove launch and landing nodes
            if length(dd) == 1
                continue
            end
            num = randperm(length(dd))[1:2]
            ad = []
            push!(ad, findfirst(x -> x == dd[num[1]], ff))
            push!(ad, findfirst(x -> x == dd[num[2]], ff))
            sort!(ad)
            ff[ad[1]:ad[2]] = reverse(ff[ad[1]:ad[2]])
            cost1 = 0
            for ii in 1:length(ff)-1
                cost1 += Du[ff[ii], ff[ii+1]]
            end
            if cost1 < L
                yy = DroneRoute[dr[i], ad[2]:-1:ad[1], u]
                DroneRoute[dr[i], ad[1]:ad[2], u] = yy
                cc = 1
            end
        end
        if cc == 1
            break
        end
    end    

    Sol["TruckRoute"] = TruckRoute
    Sol["DroneRoute"] = DroneRoute
    return Sol, flag
end
```

```julia
function swaptr1(model, Sol, flag)

    nNodes = model["nNodes"]
    ua = model["uav"]
    tru = model["tru"]
    L = model["L"]
    Du = model["Du"]
    Dt = model["Dt"]

    TruckRoute = Sol["TruckRoute"]
    DroneRoute = Sol["DroneRoute"]
    flag = 0

    tr = TruckRoute

    trr = []
    for i in 1:tru
        if count(!iszero, TruckRoute[i, :]) >= 3
            push!(trr, i)
        end
    end

    ti = rand(1:length(trr))  # Randomly select a truck route
    t = trr[ti]
    node1 = []
    node1 = vcat(node1, tr[t, 2:count(!iszero, tr[t, :])-1])

    aa = rand(1:length(node1))  # Randomly select a node from node1
    a = node1[aa]
    sa_so = findfirst(x -> x == a, tr)
    sa, so = Tuple(sa_so)
    m, n = size(tr)

    alternative = []  # Nodes available for swapping    
    alternative = vcat(alternative, vec(tr))  # ✅ Flattens `tr` into a vector


    alternative = filter(x -> x != 1 && x != 0 && x != a, alternative)

    q, w = size(DroneRoute[:, :, 1])
    truckroute = zeros(tru, n, nNodes)  # Stores possible truck routes after swap
    droneroute = zeros(q, w, ua, nNodes)  # Stores possible drone routes after swap
    swap = zeros(nNodes, 1)  # Indicates if a swap is possible for each node

    for i in alternative
        satr_sotoon = findfirst(x -> x == i, tr)
        satr, sotoon = Tuple(satr_sotoon)
        count1 = 0
        la1 = 0
        la2 = 0
        
        #added:
        ind = findall(x -> x == a, DroneRoute)
        if !isempty(ind)
            la1 = 1            
            s1, s2, s3 = getindex.(ind, 1), getindex.(ind, 2), getindex.(ind, 3)

            for tt in 1:length(s1)
                f = 0
                flightrange = DroneRoute[s1[tt], 1:count(!iszero, DroneRoute[s1[tt], :, s3[tt]]), s3[tt]]
                if s2[tt] == 1  # If `a` is a launch node
                    flightrange = flightrange[flightrange .!= DroneRoute[s1[tt], s2[tt], s3[tt]]]
                    f += Du[i, flightrange[1]]
                    for uh in 1:length(flightrange)-1
                        f += Du[flightrange[uh], flightrange[uh+1]]
                    end
                    if f <= L
                        count1 += 1
                    end
                elseif s2[tt] != 1  # If `a` is a landing node
                    flightrange = flightrange[flightrange .!= DroneRoute[s1[tt], s2[tt], s3[tt]]]
                    f += Du[flightrange[end], i]
                    for yv in 1:length(flightrange)-1
                        f += Du[flightrange[yv], flightrange[yv+1]]
                    end
                    if f <= L
                        count1 += 1
                    end
                end
            end
        end
        
        #added:
        ind = findall(x -> x == i, DroneRoute)
        if !isempty(ind)
            la2 = 1            
            i1, i2, i3 = getindex.(ind, 1), getindex.(ind, 2), getindex.(ind, 3)
            count2 = 0
            for tt in 1:length(i1)
                f = 0
                flightrange = DroneRoute[i1[tt], 1:count(!iszero, DroneRoute[i1[tt], :, i3[tt]]), i3[tt]]
                if i2[tt] == 1  # If `i` is a launch node
                    flightrange = flightrange[flightrange .!= DroneRoute[i1[tt], i2[tt], i3[tt]]]
                    f += Du[a, flightrange[1]]
                    for u in 1:length(flightrange)-1
                        f += Du[flightrange[u], flightrange[u+1]]
                    end
                    if f <= L
                        count2 += 1
                    end
                elseif i2[tt] != 1  # If `i` is a landing node
                    flightrange = flightrange[flightrange .!= DroneRoute[i1[tt], i2[tt], i3[tt]]]
                    f += Du[flightrange[end], a]
                    for y in 1:length(flightrange)-1
                        f += Du[flightrange[y], flightrange[y+1]]
                    end
                    if f <= L
                        count2 += 1
                    end
                end
            end
        end
        
        # Determine if swap is possible
        if (la1 == 1 && la2 == 1 && count1 == length(s1) && count2 == length(i1))
            swap[i, 1] = 1
            truckroute[:, :, i] = tr
            truckroute[satr, sotoon, i] = a
            truckroute[sa, so, i] = i
            droneroute[:, :, :, i] = DroneRoute
            for zz in 1:length(s1)
                droneroute[s1[zz], s2[zz], s3[zz], i] = i
            end
            for ss in 1:length(i1)
                droneroute[i1[ss], i2[ss], i3[ss], i] = a
            end
        elseif (la1 == 0 && la2 == 0)
            swap[i, 1] = 1
            truckroute[:, :, i] = TruckRoute
            truckroute[satr, sotoon, i] = a
            truckroute[sa, so, i] = i
            droneroute[:, :, :, i] = DroneRoute
        elseif (la1 == 1 && la2 == 0 && count1 == length(s1))
            swap[i, 1] = 1
            truckroute[:, :, i] = TruckRoute
            truckroute[satr, sotoon, i] = a
            truckroute[sa, so, i] = i
            droneroute[:, :, :, i] = DroneRoute
            for zz in 1:length(s1)
                droneroute[s1[zz], s2[zz], s3[zz], i] = i
            end
        elseif (la1 == 0 && la2 == 1 && count2 == length(i1))
            swap[i, 1] = 1
            truckroute[:, :, i] = TruckRoute
            truckroute[satr, sotoon, i] = a
            truckroute[sa, so, i] = i
            droneroute[:, :, :, i] = DroneRoute
            for zz in 1:length(i1)
                droneroute[i1[zz], i2[zz], i3[zz], i] = a
            end
        end
    end

    if isempty(findall(x -> x == 1, swap))
        flag = 1
        return Sol, flag
    else
        if count(!iszero, swap) == 1
            d = findall(x -> x == 1, swap)
            c = rand(1:length(d))
            b = d[c]
        else
            b = rand(findall(x -> x == 1, swap))
        end
        TruckRoute = truckroute[:, :, b]
        DroneRoute = droneroute[:, :, :, b]
    end
    
    Sol["TruckRoute"] = Int64.(TruckRoute)
    Sol["DroneRoute"] = Int64.(DroneRoute)
    return Sol, flag
end
```

```julia
using Random
using StatsBase  # Needed for sample()

function swaptruck1(model, Sol, flag)
    nNodes = model["nNodes"]
    ua = model["uav"]
    tru = model["tru"]
    L = model["L"]
    Du = model["Du"]
    Dt = model["Dt"]

    TruckRoute = Sol["TruckRoute"]
    DroneRoute = Sol["DroneRoute"]
    flag = 0

    truckroute = copy(TruckRoute)
    for i in 1:tru
        last_nonzero = findlast(!iszero, truckroute[i, :])
        if last_nonzero !== nothing
            truckroute[i, last_nonzero] = 0
        end
    end
    truckroute[:, 1] .= 0  # Remove the initial depot

    uav = Int[]
    
    #added:
    for u in 1:ua
        satr = findall(x -> x != 0, DroneRoute[:, 1, u])  # Find the launch node
        if !isempty(satr)  # Check if UAV has a flight
            push!(uav, u)            
            for in in 1:length(satr)                
                indices = findall(x -> x == DroneRoute[satr[in], 1, u], truckroute)
                if !isempty(indices)
                    (aa, bb) = Tuple(indices[1])
                    truckroute[aa, bb] = 0
                else
                    aa, bb = 0, 0  # Assign safe defaults
                    
                end

                last_index = findlast(!iszero, DroneRoute[satr[in], :, u])
                indices = findall(x -> x == DroneRoute[satr[in], last_index, u], truckroute)
                if !isempty(indices)
                    (cc, dd) = Tuple(indices[1])
                    truckroute[cc, dd] = 0
                else
                    cc, dd = 0, 0  # Assign safe defaults
                end                
                
            end
        end
    end
    nonzero = findall(!iszero, truckroute)
    if length(nonzero) < 2
        flag = 1
        return Sol, flag
    end
    
    t = sample(1:length(nonzero), 2, replace=false)
    a, b = truckroute[nonzero[t[1]]], truckroute[nonzero[t[2]]]
    
    TruckRoute[nonzero[t[1]]] = b
    TruckRoute[nonzero[t[2]]] = a    

    Sol["TruckRoute"] = TruckRoute
    Sol["DroneRoute"] = DroneRoute
    return Sol, flag
end
```

```julia
function swaptrone1(model, Sol, flag)

    nNodes = model["nNodes"]
    ua = model["uav"]
    tru = model["tru"]
    L = model["L"]
    Du = model["Du"]
    Dt = model["Dt"]
    
    TruckRoute = Sol["TruckRoute"]
    DroneRoute = Sol["DroneRoute"]
    
    flag = 0

    truckroute = copy(TruckRoute)
    for i in 1:tru
        truckroute[i, count(!iszero, truckroute[i, :])] = 0
    end
    truckroute[:, 1] .= 0

    uav = []
    
    #added:
    for u in 1:ua                
        satr = findall(!iszero, DroneRoute[:, 1, u])
        if !isempty(satr)
            push!(uav, u)            

            for in in 1:length(satr)                
                indices = findfirst(x -> x == DroneRoute[satr[in], 1, u], truckroute)                  
                if indices !== nothing  
                    aa, bb = Tuple(indices)  
                    truckroute[aa, bb] = 0
                end            
                indices = findfirst(x -> x == DroneRoute[satr[in], count(!iszero, DroneRoute[satr[in], :, u]), u], truckroute)  
                if indices !== nothing  
                    cc, dd = Tuple(indices)  
                    truckroute[cc, dd] = 0
                end                       
            end
        end
    end
    indices = findall(!iszero, truckroute)  # Finds all (row, col) positions

    if isempty(indices)
        flag = 1
        return Sol, flag
    else
        roww = [idx[1] for idx in indices]  # Extract row indices
        coll = [idx[2] for idx in indices]  # Extract column indices
    
        rnd = rand(1:length(roww))
        trex = TruckRoute[roww[rnd], coll[rnd]]
    end
    

    if isempty(uav)
        flag = 1
        return Sol, flag
    end

    ii = rand(1:length(uav))
    row = findall(!iszero, DroneRoute[:, 1, uav[ii]])
    ro = rand(1:length(row))
    launch = DroneRoute[row[ro], 1:count(!iszero, DroneRoute[row[ro], :, uav[ii]]), uav[ii]]
    alternative = copy(launch)
    deleteat!(alternative, 1)
    deleteat!(alternative, length(alternative))

    for i in 1:length(alternative)
        sum = 0        
        sotoon = findfirst(x -> x == alternative[i], launch)         
        sum += Du[DroneRoute[row[ro], sotoon-1, uav[ii]], trex] + Du[trex, DroneRoute[row[ro], sotoon+1, uav[ii]]]
        for j in 1:length(launch)-1
            if j == sotoon || j+1 == sotoon
                continue
            end
            sum += Du[DroneRoute[row[ro], j, uav[ii]], DroneRoute[row[ro], j+1, uav[ii]]]
        end

        if sum <= L
            TruckRoute[roww[rnd], coll[rnd]] = DroneRoute[row[ro], sotoon, uav[ii]]
            DroneRoute[row[ro], sotoon, uav[ii]] = trex
            break
        end
    end
    
    Sol["TruckRoute"] = TruckRoute
    Sol["DroneRoute"] = DroneRoute
    return Sol, flag
end
```

```julia
using LinearAlgebra

function CoolingRate(Current, PRIM)
    Peak = [1.025, 0.9, 0.8]  # Define the Peak vector
    BestSol = Current  # Assign Current to BestSol

    # Calculate DELTA_C - element-wise operations
    DELTA_C = (BestSol .- PRIM["Cost"]) ./ BestSol
    alpha = 1 .- 0.99 .* exp.(-5 .* DELTA_C.^2)  # Note the dots for element-wise operations
    DELTA_C = DELTA_C ./ alpha

    # Initialize variables
    better = 0.0
    nochange = 0.0
    worse = 0.0

    # Determine the values of worse, nochange, and better based on DELTA_C
    # Note: DELTA_C is now a vector, so we need to handle each element
    # This assumes DELTA_C is a single-element vector - if not, you'll need to decide how to handle it
    delta = DELTA_C[1]  # Taking first element if it's a single-element vector
    
    if delta <= -1
        worse = 1.0
    elseif delta > -1 && delta <= 0
        worse = -delta
        nochange = 1 + delta
    elseif delta > 0 && delta <= 1
        nochange = 1 - delta
        better = delta
    elseif delta > 1
        better = 1.0
    end

    # Create the Mu vector
    Mu = [worse; nochange; better]

    # Calculate CR as the dot product of Peak and Mu
    CR = dot(Peak, Mu)

    return CR
end
```
